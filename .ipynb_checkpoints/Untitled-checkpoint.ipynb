```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
datab = db.companies.find()
df = pd.DataFrame(datab)
display(df.head(1)) # DataFrame original
df.shape # Comprobar filas y columnas del DF original
```

```python
# Se empieza a limpiar el DF con queries por pymongo
```

```python
mongo_office= db.companies.find({
    "offices":{"$not":{"$size":0}}},
    {"name":1, "offices":1, "category_code":1, "founded_year":1,"description":1,
     "deadpooled_year":1, "number_of_employees":1, "total_money_raised":1}) 
# Empresas con una o más oficinas
```

```python
df_office = pd.DataFrame(mongo_office)
display(df_office)
```

```python
df_office.isna().sum()
```

```python

def cleaning(data):
#funcion para coger las lats y longs de la columna office
    data = data['offices']

    for i in data:
        if i['latitude'] and i['longitude']:
            return {
        "lat": i['latitude'],
        "lng": i['longitude']
        }

new_df_coords = df_office[["offices"]].apply(cleaning, result_type="expand", axis=1)
display(new_df_coords)
```

```python
df_clean = pd.concat([new_df_coords,df_office], axis=1)[["name","category_code", "deadpooled_year", 
                                                            "founded_year","description", "lat", 
                                                         "lng", "number_of_employees", "total_money_raised"]]


display(df_clean)
```

```python
df_clean.dropna(subset=["lat", "lng"], inplace=True)
```

```python
display(df_clean)
```

```python
df_clean.isna().sum()
```

```python
df_clean.dropna(subset=["lat", "lng"], inplace=True)
```

```python
df_clean = df_clean[df_clean['deadpooled_year'].isna()]
```

```python
display(df_clean)
```

```python
df_clean = df_clean.drop(df_clean[df_clean["number_of_employees"]== 0].index)
```
